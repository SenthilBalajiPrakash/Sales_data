**Data Distribution**

2017 Data

```python
s17.columns
```

```python
s17_num_cols = ['Qty Itens', 'Unit Price', 'Net Weight']

plt.figure(figsize=(12, 6))
for i, col in enumerate(s17_num_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data=s17, x=col, kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')


    plt.xlim(s17[col].quantile(0.05), s17[col].quantile(0.5))
plt.tight_layout()
plt.show()
```

```python
for col in s17_num_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=s17[col])

    # Calculate quartiles
    q1 = s17[col].quantile(0.25)
    q2 = s17[col].median()
    q3 = s17[col].quantile(0.75)

    # Write quartiles inside the figure


    plt.ylim(s17[col].quantile(0.05), s17[col].quantile(0.95))

    plt.title(f'Box plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Values')
    plt.show()
```

```python

outlier_counts = {}


for col in s17_num_cols:
    q1 = s17[col].quantile(0.25)
    q3 = s17[col].quantile(0.75)
    iqr = q3 - q1

    lb_iqr = q1 - 1.5 * iqr
    ub_iqr = q3 + 1.5 * iqr

    outliers = s17[(s17[col] < lb_iqr) | (s17[col] > ub_iqr)]
    outlier_counts[col] = len(outliers)

for col, count in outlier_counts.items():
    print(f"Outlier count in '{col}': {count}")
```
